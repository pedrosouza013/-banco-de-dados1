# AV2-01/7: técnicas de amostragem   

Neste notebook, utiliza-se um dataset fictício que foi gerado por Inteligência Artificial (Copilot) para um exercício prático de Análise de Sentimento na matéria de IA Generativa de Organizações. Contém 30 registros baseados em avaliações reais sobre a seguradora fictícia "Seguros Río de Plata", divididos entre opiniões de clientes, funcionários e postagens em redes sociais.  

A estrutura do arquivo CSV inclui campos como identificador único (`id`), origem do depoimento (`fuente`), o texto completo (`comentario`), data, nome da empresa e uma nota de 1 a 5 (`rating`), esta última aplicada apenas aos clientes. Com dados que variam entre elogios por atendimentos rápidos e reclamações de aumentos abusivos, o conjunto é ideal para testar ou treinar modelos de processamento de linguagem natural (NLP), utilizando bibliotecas como VADER, TextBlob ou Transformers para identificar nuances e sentimentos positivos, negativos ou neutros.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregando o dataset da seguradora
dados = pd.read_csv('simulacion_comentarios.csv')

# Exibe amostras do dataset
dados.head()
```

```python
# Amostragem Aleatória Simples
tamanho_amostra = 15
amostra_aleatoria = dados.sample(n=tamanho_amostra, random_state=1)

# Amostragem Estratificada (Usando a coluna 'fuente')
estratos = dados['fuente'].unique()
amostra_estratificada = dados.groupby('fuente', group_keys=False).apply(
    lambda x: x.sample(min(len(x), tamanho_amostra // len(estratos)), random_state=1), include_groups=False
)

# Amostragem Sistemática
intervalo = len(dados) // tamanho_amostra
indices_sistematicos = np.arange(0, len(dados), intervalo)
amostra_sistematica = dados.iloc[indices_sistematicos[:tamanho_amostra]]
```

```python

# Visualização das distribuições de Rating (removendo nulos para o gráfico)
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

axs[0].hist(amostra_aleatoria['rating'].dropna(), bins=10, color='blue', alpha=0.7)
axs[0].set_title('Amostragem Aleatória Simples - Distribuição de Ratings')

axs[1].hist(amostra_estratificada['rating'].dropna(), bins=10, color='green', alpha=0.7)
axs[1].set_title('Amostragem Estratificada - Distribuição de Ratings')

axs[2].hist(amostra_sistematica['rating'].dropna(), bins=10, color='red', alpha=0.7)
axs[2].set_title('Amostragem Sistemática - Distribuição de Ratings')

plt.tight_layout()
plt.show()
```
